```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
pca = PCA(n_components= 2)
X_pca = pca.fit_transform(X)
```

```python
first_component = X_pca[:, 0]
second_component = X_pca[:, 1]
```

### Clustering with kmeans

```python
from sklearn.cluster import KMeans
```

```python
ewscc = []
```

```python
for i in range(1, 150):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit_transform(X)
    ewscc.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 150), ewscc)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
accuracy_list = []
```

```python
for i in range(2, 150):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit_transform(X)
    a = silhouette_score(X, kmeans.labels_)
    print(i, a)
    accuracy_list.append(a)
```

```python
kmeans = KMeans(n_clusters= 147, random_state= 20)
```

```python
kmeans.fit(X)
```

```python
plt.scatter(first_component, second_component, c= kmeans.labels_, alpha = 0.7)
plt.show()
```
