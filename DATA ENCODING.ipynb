## data encoding
data encoding refers to the converting of categorical variable into numeriacal representation that can understerstood by machine learning algorutims


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

## LABEL ENCODING

```python
df.species.unique()
```

```python
from sklearn import preprocessing
```

```python
# intitiallize label encoder
label_enc = preprocessing.LabelEncoder()
```

```python
df['Species Encoded'] = label_enc.fit_transform(df.species)
```

```python
df.head()
```

```python
df['Species Encoded'] . unique()
```

## label encoding with pandas
