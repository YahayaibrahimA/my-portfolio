```python
# fetch all movies rated PG-13
h =df['Rating'] == 'PG-13'
df[h]
```

```python
# FETCH ALL NOT RATED MOVIES WITH AN IMDB SCORE ABOVE 6.
k= df['Rating'] == 'Not Rated'
n = df['IMDb Score (1-10)'] > 6.0
df[k|n]
```

```python
# RATING
g = df['Rating'] == 'G'
df[g]
```

```python
# FETCH ALL G-RATED MOVIES WITH GROSS REVENUE ABOVE $1,000,000
r = df['Gross Revenue'] > 1000000.0
df[r]
```

```python
# SCORE AND REVIEW FETCH
df[g * r]
```

```python
# FETCH ALL MOVIES WITH AN IMDB SCORE OF 8.0 HIGHER.
x = df['IMDb Score (1-10)'] >= 8.0
df[x]
```

```python
## FETCH ALL MOVIES WITH TOTAL REVIEWS GREATER THAN 200
u = df['Total Reviews'] > 200.0
df[u]
```

```python
## fetch all movies with an imdb score below
v = df['IMDb Score (1-10)'] < 5.0
df[v][['Genre', 'Director Name']]
```

```python
# FETCH ALL MOVIES WHERE THE LEAD ACTOR IS KAREN TROUT
y = df['Lead Actor'] == 'Karen Trout'
df[y]
```

```python
# FETCH ALL MOVIES DIRECTOR BY TODD WAFFLE.
df[df['Director Name'] == 'Todd Waffle']
```

```python
# FETCH ALL MOVIES WHERE THE LEAD ACTOR FB LIKE IS HIGHER THEN DIRECTOR FB LIKES.
u = df['Lead Actor FB Likes'] 
b = df['Director FB Likes']
df[u > b]
```
