### Label Encoding Species Column

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Label_enc = LabelEncoder()
```

```python
Label_enc.fit_transform(df.species)
```

```python
df['species'] = Label_enc.fit_transform(df.species)
```

df.head()

```python
x = df.drop('species', axis = 1)
y = df['species']
```

```python
x.head()
```

```python
y.head()
```

### Split x and y into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
# Split x and y into training, use 80% as train set and 20% a test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random _state =)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

### Model Training

kneighbor classifier

```python
# kneighbor classifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initialize the knn classifier
knn = KNeighborsClassifier(n_neighbors = 10)
```

```python
knn.fit(x_train, y_train)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix 
```

```python
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
