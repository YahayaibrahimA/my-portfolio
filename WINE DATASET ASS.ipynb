## Feature selection using selectkbest

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import chi2
```

```python
x = df.drop('quality', axis = 1)
y = df.quality
```

```python
# select top 7 feature
# intitialze the selectkbeat
select = SelectKBest(chi2, k=7)
```

```python
X_new = select.fit_transform(x, y)
```

```python
x.columns[select.get_support()]
```

```python
x = x[['fixed acidity', 'volatile acidity', 'citric acid',
       'free sulfur dioxide', 'total sulfur dioxide', 'alcohol', 'Id']]
```

```python
x.head()
```

## Checking for class imbalance in Output(Y) variable

### . Oversampling
 ##   .Undersampling
 ##    . Class Weight

```python
class_count = df.groupby('quality').agg({'Id' : 'count'}).reset_index()
class_count
```

```python
sns.barplot(data=class_count, x= class_count['quality'], y= class_count['Id'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x, y, random_state= 25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initital the model
Knn =  KNeighborsClassifier(n_neighbors= 82)
```

```python
Knn.fit(x_train, y_train)
```

```python
y_pred = Knn.predict(x_test)
```

## MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
n_neighbors= []
accuracy_list = []
```

```python
for i in range(1, 201):
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    n_neighbors.append(i)
    a = accuracy_score(y_test, y_pred)
    accuracy_list.append(a)
```

```python
plt.plot(n_neighbors, accuracy_list)
plt.show()
```
